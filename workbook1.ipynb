The simplest way of creating a Figure with an Axes is using pyplot.subplots. We can then use Axes.plot to draw some data on the Axes, and show to display the figure:

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4,5,6],[1,3,4,6,2,1])
plt.show()

#this is one figure and one axes
```

Figure keeps all the axes together.

```python
fig = plt.figure()
fig, ax = plt.subplots()
fig, axs = plt.subplots(2,3)
```

```python
# a figure with one Axes on the left, and two on the right:
fig, axs = plt.subplot_mosaic([['left', 'right_top'],
                               ['left', 'right_bottom']])
```

```python
fig = plt.figure(figsize=(2, 2), facecolor='lightskyblue', layout='constrained')

fig.suptitle('Figure')
ax = fig.add_subplot()
ax.set_title('Axes', loc='left', fontstyle='oblique', fontsize='medium')
```
